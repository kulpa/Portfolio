```python
from pandas import DataFrame, read_csv
import matplotlib.pyplot as plt
import pandas as pd 
import sys 

%matplotlib inline

"""This file does the autoregression of the crude oil"""

#Crude oil analysis
OIL_LOCATION = r'../data/CrudeOilPrice.csv'
DF_OIL = pd.read_csv(OIL_LOCATION) 

DF_OIL["DATE"] = pd.to_datetime(DF_OIL["DATE"], '%Y-%m-%d')

DF_OIL.info()
DF_OIL.plot(title="Crude Oil Price", x="DATE", y="VALUE")
```

```python
DF_OIL.describe()
```

```python
DF_OIL_LATER = DF_OIL[1:]
DF_OIL_LATER.index = [range(len(DF_OIL_LATER))]
DF_OIL_LATER
```

```python
DF_OIL_EALIER = DF_OIL[:356]
DF_OIL_EALIER
```

```python
Autoregression = pd.concat([DF_OIL_EALIER["VALUE"], DF_OIL_LATER["VALUE"]], axis=1)
Autoregression.columns = ['Early', 'Later']
Autoregression.plot(kind='scatter' , title = "Autoregression",x="Early",y="Later")

Autoregression.corr()
```

```python
model = pd.ols(x = DF_OIL_EALIER["VALUE"], y = DF_OIL_LATER["VALUE"])
model
```
