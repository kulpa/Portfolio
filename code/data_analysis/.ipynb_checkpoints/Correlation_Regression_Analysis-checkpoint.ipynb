```python
from pandas import DataFrame, read_csv
import matplotlib.pyplot as plt
import pandas as pd 
import sys 

%matplotlib inline

"""This file analysises the correlation between crude oil price to BDI, WS,Container fare"""

#Crude oil analysis
OIL_LOCATION = r'../data/Oil_Price.csv'
DF_OIL = pd.read_csv(OIL_LOCATION) 

DF_OIL["Date"] = pd.to_datetime(DF_OIL["Date"], '%Y/%m/%d')

DF_OIL.info()
DF_OIL.plot(title="Crude Oil Price", x="Date", y="Oil_Value")
```

```python
#BDI analysis
BDI_LOCATION = r'../data/BDI.csv'
DF_BDI = pd.read_csv(BDI_LOCATION) 

DF_BDI["Date"] = pd.to_datetime(DF_BDI["Date"], '%Y/%m/%d')

DF_BDI.plot(title="BDI", x="Date", y="BDI_Value")
```

```python
#WS analysis
WS_LOCATION = r'../data/WS.csv'
DF_WS = pd.read_csv(WS_LOCATION) 

DF_WS["Date"] = pd.to_datetime(DF_WS["Date"], '%Y/%m/%d')

DF_WS.plot(title="WS", x="Date", y="WS_Value")
```

```python
"""Calculate the correlation with Oil Price and BDI"""

#the same date and number of column with BDI
BDI_same_DF_OIL = DF_OIL[74:]

BDI_same_DF_OIL.index = [range(len(BDI_same_DF_OIL))]

print "比較したデータは" + str(len(BDI_same_DF_OIL)) + "ヶ月です。"

Oil_and_BDI = pd.concat([BDI_same_DF_OIL["Oil_Value"], DF_BDI["BDI_Value"]], axis=1)
Oil_and_BDI.plot(kind='scatter', title = "Oil price and BDI",x="Oil_Value",y="BDI_Value")

#Oil_and_BDI.describe()
Oil_and_BDI.corr()
```

```python
"""Calculate the regression of BDI from oil price"""
model = pd.ols(x = Oil_and_BDI["Oil_Value"], y = Oil_and_BDI["BDI_Value"])
model
# model.summary_as_matrix
# BDI = -15 * Oil price + 2971
```

```python
"""Calculate the correlation with Oil Price and WS"""

#the same date and number of column with WS
WS_same_DF_OIL = DF_OIL[:-19]

WS_same_DF_OIL.index = [range(len(WS_same_DF_OIL))]

print "比較したデータは" + str(len(WS_same_DF_OIL)) + "ヶ月です。"

Oil_and_WS = pd.concat([WS_same_DF_OIL["Oil_Value"], DF_WS["WS_Value"]], axis=1)
Oil_and_WS.plot(kind='scatter', title = "Oil price and WS",x="Oil_Value",y="WS_Value")

#Oil_and_WS.describe()
Oil_and_WS.corr()
```

```python
"""Calculate the regression of WS from oil price"""
model = pd.ols(x = Oil_and_WS["Oil_Value"], y = Oil_and_WS["WS_Value"])
model
# model.summary_as_matrix
# WS = -1.01 * Oil price + 152
```
