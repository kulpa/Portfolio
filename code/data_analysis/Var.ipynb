```python
#BSI analysis
BSI_LOCATION = r'../data/LLOYDS-BSI.csv'
DF_BSI = pd.read_csv(BSI_LOCATION) 

DF_BSI["Date"] = pd.to_datetime(DF_BSI["Date"], '%Y-%m-%d')

DF_BSI.plot(title="BSI", x="Date", y="BSI_Value")
```

```python
# DF_BCI_E: from 2009/3/31 - 2015/9/30
DF_BCI_E = DF_BCI[1:]
DF_BCI_E = DF_BCI_E.iloc[::-1]
DF_BCI_E.index = [range(len(DF_BCI_E))]

# DF_BCI_L: from 2009/4/30 - 2015/10/31
DF_BCI_L = DF_BCI[:79]
DF_BCI_L = DF_BCI_L.iloc[::-1]
DF_BCI_L.index = [range(len(DF_BCI_L))]
```

```python
# DF_BPI_E: from 2009/3/31 - 2015/9/30
DF_BPI_E = DF_BPI[1:]
DF_BPI_E = DF_BPI_E.iloc[::-1]
DF_BPI_E.index = [range(len(DF_BPI_E))]

# DF_BPI_L: from 2009/4/30 - 2015/10/31
DF_BPI_L = DF_BPI[:79]
DF_BPI_L = DF_BPI_L.iloc[::-1]
DF_BPI_L.index = [range(len(DF_BPI_L))]
```

```python
# multiregression of BCI
import statsmodels.formula.api as sm
df = pd.DataFrame({"BCIt": DF_BCI_L["BCI_Value"], "BCIt1": DF_BCI_E["BCI_Value"], "BPIt1": DF_BPI_E["BPI_Value"], "Oil": DF_OIL_LATER["VALUE"]})
result = sm.ols(formula="BCIt ~ BCIt1 + BPIt1 + Oil", data=df).fit()
print result.summary()
```

```python
# multiregression of BPI
import statsmodels.formula.api as sm
df = pd.DataFrame({"BPIt": DF_BPI_L["BPI_Value"], "BPIt1": DF_BPI_E["BPI_Value"],"BCIt1": DF_BCI_E["BCI_Value"], "Oil": DF_OIL_LATER["VALUE"]})
result = sm.ols(formula="BPIt ~ BPIt1 + BCIt1 + Oil", data=df).fit()
print result.summary()
```
